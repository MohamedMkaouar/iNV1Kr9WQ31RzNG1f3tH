```python
import pickle
import pandas as pd
import numpy as np
from tqdm import tqdm

import warnings
warnings.filterwarnings("ignore")
```

```python
health_topics = pickle.load(open('../resources/health_topics.pickle','rb'))
medical_tests = pickle.load(open('../resources/medical_tests.pickle','rb'))
```

```python
health_topics
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
corpus = list(health_topics.values())+list(np.array(list(medical_tests.values()))[:,0])
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(corpus)
df = pd.DataFrame(X.T.toarray(),columns = list(health_topics.keys())+list(medical_tests.keys()))
df
```

```python

score = []
for c in df.columns:
    score.append(cosine_similarity(df[["Appendicitis"]].to_numpy().reshape(1, -1),df[[c]].to_numpy().reshape(1, -1))[0,0])
best_matches_covid_vaccines = pd.DataFrame( {'health topic': df.columns,'score':score})
```

```python

best_matches_covid_vaccines.sort_values(by="score",ascending=False)
```

```python
vector = vectorizer.transform(["chest pain obese dyspnea"]).T.toarray()
```

```python
score = []
for i,c in enumerate(df.columns):
    if i >= 1909:
        score.append(cosine_similarity(vector.reshape(1, -1),df[[c]].to_numpy().reshape(1, -1))[0,0])
best_matches = pd.DataFrame( {'health topic': list(df.columns)[1909:],'score':score}).sort_values(by="score",ascending=False)
```
